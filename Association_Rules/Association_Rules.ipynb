```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('book.csv')
df.head()
```

```python
df.isna().sum()
```

```python
df.describe()
```

```python
df.info()
```

```python
df[df.duplicated()]
```

```python
df.drop_duplicates(inplace= True)
```

```python
df.duplicated().sum()
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
frequent_items = apriori(df,min_support= 0.2,use_colnames=True)
frequent_items
```

* Apriori Algorithm gives us itemsets which are important

```python
model = association_rules(frequent_items,metric='lift')
model
```

```python
rules1 = association_rules(frequent_items, metric="confidence", min_threshold=0.6)
rules1.head()
```

```python
#Visualizations

#Support vs Confidence

plt.scatter(rules1['support'], rules1['confidence'], alpha=0.5)
plt.xlabel('support')
plt.ylabel('confidence')
plt.title('Support vs Confidence')
plt.show()
```

```python
#Support vs Lift

plt.scatter(rules1['support'], rules1['lift'], alpha=0.5)
plt.xlabel('support')
plt.ylabel('lift')
plt.title('Support vs Lift')
plt.show()
```

```python
# Lift vs Confidence

fit = np.polyfit(rules1['lift'], rules1['confidence'], 1)
fit_fn = np.poly1d(fit)
plt.plot(rules1['lift'], rules1['confidence'], 'yo', rules1['lift'], 
 fit_fn(rules1['lift']))
```
